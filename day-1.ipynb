# Part 1

```python
# Open and read the file
numbers1 = []
numbers2 = []

with open('1-chiefs-lists', 'r') as file:
    for line in file:
        # Split each line on whitespace and convert to integers
        num1, num2 = map(int, line.strip().split())
        numbers1.append(num1)
        numbers2.append(num2)

print(len(numbers1))
print(len(numbers2))
```

```python
numbers1.sort()
numbers2.sort()

differences = [max(numbers1[i], numbers2[i]) - min(numbers1[i], numbers2[i]) for i in range(len(numbers1))]

sum_of_differences = sum(differences)
print(sum_of_differences)
```

# Part 2

```python
from collections import Counter
number_counts = Counter(numbers2)
print(f"Occurrences in numbers2: {dict(number_counts)}")
```

```python
similarity_score = 0

for number in numbers1:
    similarity_score += number_counts[number] * number

print(similarity_score)
```
